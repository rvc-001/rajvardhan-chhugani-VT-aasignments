```python
marks = [78, 92, 85, 67, 95, 88, 91, 73, 99, 84, 89, 93]

highest = max(marks)
lowest = min(marks)
average = sum(marks) / len(marks)

above_avg = len([m for m in marks if m > average])
above_90 = len([m for m in marks if m > 90])

top10 = sorted(marks, reverse=True)[:10]

print("Highest:", highest)
print("Lowest:", lowest)
print("Average:", average)
print("Above Average:", above_avg)
print("Above 90:", above_90)
print("Top 10:", top10)
```

```python
spending = [1200, 4500, 2300, 8900, 5600, 1000, 7800, 9200, 3300, 6700]

top5 = sorted(spending, reverse=True)[:5]
bottom5 = sorted(spending)[:5]

s = sorted(spending)
n = len(s)
median = (s[n//2-1] + s[n//2]) / 2

revenue = sum(spending)

target = revenue * 0.8
running = 0
customers = 0

for amount in sorted(spending, reverse=True):
    running += amount
    customers += 1
    if running >= target:
        break

print(top5)
print(bottom5)
print("Median:", median)
print("Revenue:", revenue)
print("Customers for 80% revenue:", customers)
```

```python
data = [10, 20, None, 50, '', 90, -5, 'NA', 100]

invalid = [x for x in data if x in [None, '', 'NA']]

valid = [x for x in data if isinstance(x, int) and x >= 0]

mean = sum(valid) / len(valid)

cleaned = []

for x in data:
    if x in [None, '', 'NA']:
        cleaned.append(mean)
    elif isinstance(x, int) and x >= 0:
        cleaned.append(x)

print("Invalid Count:", len(invalid))
print("Mean:", mean)
print("Cleaned:", cleaned)
```

```python
population = [500000]

for i in range(100):
    population.append(population[-1] * 1.03)

growth = []

for i in range(1, len(population)):
    growth.append(population[i] - population[i-1])

max_growth = max(growth)
year = growth.index(max_growth) + 1

for i, p in enumerate(population):
    if p > 1000000:
        first_year = i
        break

print("Max Growth Year:", year)
print("First Year > 1M:", first_year)
```

```python
from collections import Counter

ratings = [5,4,5,3,4,5,2,5]

avg = sum(ratings)/len(ratings)

freq = Counter(ratings)

print("Average:", avg)
print("Most Frequent:", freq.most_common(1))

print("Distribution:", freq)

high = len([x for x in ratings if x >= 4])

print("Highly Rated:", high)
```

```python
students = [
    ['Rahul',85,90,88],
    ['Amit',72,80,75],
    ['Priya',95,91,93]
]

topper = ""
highest = 0

for s in students:
    total = sum(s[1:])
    percentage = total / 3

    if total > highest:
        highest = total
        topper = s[0]

    if percentage >= 90:
        grade = "A"
    elif percentage >= 75:
        grade = "B"
    else:
        grade = "C"

    print(s[0], total, percentage, grade)

print("Topper:", topper)

class_avg = sum(sum(s[1:])/3 for s in students)/len(students)

print("Class Average:", class_avg)

subject1 = max(students, key=lambda x:x[1])
subject2 = max(students, key=lambda x:x[2])
subject3 = max(students, key=lambda x:x[3])

print("Highest Subject 1:", subject1[0])
print("Highest Subject 2:", subject2[0])
print("Highest Subject 3:", subject3[0])
```
